```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel("/content/cutoff_predict.xlsx")
df.head(10)
```

```python
df["cutoff"].astype(float)
df.dtypes
```

```python
df.isnull().sum()
```

```python
one_hot_encoded_data = pd.get_dummies(df, columns = ['quota', 'dept','difficulty_level','job_demand'])
one_hot_encoded_data
```

```python
df.shape
```

```python
x = one_hot_encoded_data.drop(columns=['cutoff'])
y = one_hot_encoded_data['cutoff']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8,random_state=0)
```

APPLYING LINEAR REGRESSION :


```python
from sklearn.linear_model import LinearRegression

linear_regression_model = LinearRegression()
linear_regression_model.fit(x_train, y_train)
y_pred = linear_regression_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2_lin = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_lin}')
```

APPLYING DECISIONTREE REGRESSOR :

```python
from sklearn.tree import DecisionTreeRegressor

decisionTree_regressor_model = DecisionTreeRegressor()

decisionTree_regressor_model.fit(x_train, y_train)

y_pred =  decisionTree_regressor_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)
r2_des = r2_score(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_des}')
print(f"Root Mean Squared Error: {rmse}")
```

APPLYING SVR (SUPPORT VECTOR REGRESSOR) :

```python
from sklearn.svm import SVR

svr_model = SVR(kernel='linear')

svr_model.fit(x_train, y_train)

y_pred = svr_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

r2_svm = r2_score(y_test, y_pred)

rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_svm}')
print(f"Root Mean Squared Error: {rmse}")
```

APPLYING RIDGE REGRESSOR :

```python
from sklearn.linear_model import Ridge

ridge_regressor = Ridge(alpha=1.0)

ridge_regressor.fit(x_train, y_train)


y_pred = ridge_regressor.predict(x_test)

r2_gra = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_gra}')
print(f"Root Mean Squared Error: {rmse}")
```

APPLYING GRADIENT BOOST REGRESSOR :

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)

r2_gra = r2_score(y_test, y_pred)

rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_gra}')
print(f"Root Mean Squared Error: {rmse}")
```

RANDOM FOREST REGRESSOR :

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

r2_ran = r2_score(y_test, y_pred)

rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_ran }')
print(f"Root Mean Squared Error: {rmse}")
```

XGBOOST REGRESSOR :

```python
import xgboost as xgb
import joblib

# Create and train your XGBoost model
xgb_regressor = xgb.XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    objective='reg:squarederror'
)

xgb_regressor.fit(x_train, y_train)

# Save the XGBoost model as a .pkl file
joblib.dump(xgb_regressor, 'xgb_regressor_model.pkl')

# Make predictions
y_pred = xgb_regressor.predict(x_test)
mse = mean_squared_error(y_test, y_pred)

r2_gra = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2) Score: {r2_gra}')
print(f"Root Mean Squared Error: {rmse}")
```
